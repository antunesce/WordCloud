```
# importar os pacotes necessários
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import unidecode
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

# importar o arquivo csv em um df
df = pd.read_csv('Title2.csv', encoding='Latin-1')
```

```
# eliminar as colunas com valores ausentes
summary = df.dropna(subset=['summary'], axis=0)['summary']
```

```
# concatenar as palavras
all_summary = " ".join(s for s in summary)

# ver quantidade de palavras
print("Quantidade de Palavras: {}".format(len(all_summary)))
```

```
# lista de stopword
stopwords = set(STOPWORDS)
stopwords.update(["forest", "tropical", "dry", "forests","plant"]) # Palavras que serão excluidas

# gerar uma wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="white", # Cor do fundo
                      width=1600, height=800).generate(all_summary)

# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,6)) # Tamanho da fonte das palavras
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud);
wordcloud.to_file("tropical_dry_forest_2020_2016.png") # Nome do arquivo de imagem que será gerado
```
